```python
dff['Age']=dff['Age'].fillna(dff['Age'].mean())
```

```python
dff['Fare']=dff['Fare'].fillna(dff['Fare'].mean())
```

```python
dff.drop(['Cabin'],axis=1,inplace=True)
```

```python
dff.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
dff.isnull().sum()
```

```python
dum=pd.get_dummies(dff.Sex)
```

```python
dfmerg=pd.concat([dff,dum],axis=1)
```

```python
dfmerg=dfmerg.drop(['Sex','male'],axis=1)
```

```python
dum2=pd.get_dummies(dfmerg.Embarked)
```

```python
dfmerg2=pd.concat([dfmerg,dum2],axis=1)
```

```python
dfmerg2=dfmerg2.drop(['Embarked','C'],axis=1)
```
